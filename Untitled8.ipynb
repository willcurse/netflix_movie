# 3.1: What is the distribution of content types (Movies vs. TV Shows)?

```python

movie_count = data[data['type'] == 'Movie'].shape[0]
tv_show_count = data[data['type'] == 'TV Show'].shape[0]
total_count = data['type'].count()

print(f'Total movies are {movie_count}')
print(f'Total TV shows are {tv_show_count}')
print(f'Total content count is {total_count}')

# Prepare the data for plotting
content_counts = pd.DataFrame({
    'Content Type': ['Movies', 'TV Shows'],
    'Count': [movie_count, tv_show_count]
})

# Create a barplot
plt.figure(figsize=(10,6))
sns.barplot(x='Content Type', y='Count', data=content_counts)
plt.title('Distribution of Content Types (Movies vs. TV Shows)')
plt.show()
```

# 3.2. How has the number of titles added to Netflix changed over time?

```python


# Ensure 'date_added' is in datetime format and extract the year
data['date_added'] = pd.to_datetime(data['date_added'], errors='coerce')
data['year'] = pd.DatetimeIndex(data['date_added']).year

# Group by year and count the number of titles added
titles_per_year = data.groupby('year').size().reset_index(name='count')

# Plot the number of titles added over time
plt.figure(figsize=(10,6))
sns.lineplot(x='year', y='count', data=titles_per_year)
plt.title('How the Number of Titles Added to Netflix Changed Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.show()
```

# 3.3: What are the top 10 countries producing the most content on Netflix?

```python
top_10_countries=data['country'].value_counts().head(10).reset_index()
top_10_countries.columns=['country','count']

plt.figure(figsize=(10,6))
sns.barplot(x='country',y='count',data=top_10_countries)
plt.title('The top 10 countries producing the most content on Netflix?')
plt.show()
```

# 3.6: How many seasons do TV Shows typically have on Netflix?

# 3.8:  What is the relationship between content rating and release year?

```python
data.head()
```

```python
data['release_year'] = pd.to_numeric(data['release_year'], errors='coerce')
data['rating'] = data['rating'].astype(str) 
group_data = data.groupby(['release_year', 'rating']).size().reset_index(name='count')

plt.figure(figsize=(12, 8))
pivot_table = group_data.pivot("release_year", "rating", "count")
sns.heatmap(pivot_table, cmap='viridis', annot=True, fmt='g', cbar_kws={'label': 'Number of Titles'})
plt.title('Relationship Between Content Rating and Release Year')
plt.xlabel('Content Rating')
plt.ylabel('Release Year')
plt.xticks(rotation=45)
plt.show()

sns.histplot(x='release_year',y='rating',data=group_data)
plt.title('Relationship Between Content Rating and Release Year')
plt.xlabel('Content Rating')
plt.ylabel('Release Year')
plt.show()
```
